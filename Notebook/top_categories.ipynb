```python
import pandas as pd
```

```python
titles = pd.read_csv("../content/cleaned_movie_db.csv")
```

```python
credits = pd.read_csv("../content/cleaned_credits.csv")
```

```python
titles = pd.DataFrame(titles)
credits = pd.DataFrame(credits)
```

```python
imdb_score = titles[titles['imdb_score']>=8.0]
imdb_score.head()
```

```python
imdb_sort = imdb_score.sort_values(['imdb_score'], ascending=False)
imdb_sort.head(20)
```

```python
g = imdb_score["age_certification"].str.contains('G').sum()
pg = imdb_score["age_certification"].str.contains('PG').sum()
pg13 = imdb_score["age_certification"].str.contains('PG-13').sum()
r = imdb_score["age_certification"].str.contains('R').sum()
print (g, pg, pg13, r)
```

```python
average_runtime = imdb_score["runtime"].mean()
average_runtime
```
